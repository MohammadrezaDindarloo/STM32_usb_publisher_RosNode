```python
import serial
import time
import struct
import pandas as pd
import numpy as np
```

```python
port = serial.Serial('COM39',115200,timeout=1)
```

```python
#Functions for setting the camera parameters
def camera_set_analog_gain(port_object,val):
    port_object.write(''.join(['G2 ',str(val),'\r']).encode('utf-8'))
def camera_set_digital_gain(port_object,val):
    port_object.write(''.join(['G3 ',str(val),'\r']).encode('utf-8'))
def camera_set_exposure(port_object,val):
    port_object.write(''.join(['G1 ',str(val),'\r']).encode('utf-8'))
```

```python
camera_set_analog_gain(port,30)
camera_set_digital_gain(port,12)
camera_set_exposure(port,480)
```

```python
port.flush()
data_row=[]
while True:
    t=time.time();
    data=port.read_until( terminator=''.join(['abc\n']).encode('utf-8'))
    if(len(data)==76):
        ax,ay,az,wx,wy,wz,mx,my,mz,imu_ts,camera_ts,toggling_bits,\
        rc_cmd1,rc_cmd2,rc_cmd3,rc_cmd4,rc_cmd5,rc_cmd6,rc_cmd7,rc_cmd8=struct.unpack('3h3h3h2qi8I',data)
        data_row.append([imu_ts,ax,ay,az,wx,wy,wz,mx,my,mz])
        print(rc_cmd3)
#         print(toggling_bits)
#         print(1/(t-time.time()+0.00001))
#         print(imu_ts)
```

```python
data=np.array(data_row).reshape(-1,10)
columns=['ts','ax','ay','az','gx','gy','gz','hx','hy','hz']
df=pd.DataFrame(data,columns=columns)
df['gx'].plot()
df['gy'].plot()
df['gz'].plot()
```
